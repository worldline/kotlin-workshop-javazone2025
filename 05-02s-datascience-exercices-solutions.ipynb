10. For the most-ordered item (by taking into account the quantity), how many items were ordered?


```kotlin
val chipoGroupByName = chipo.groupBy { item_name }.aggregate { sum(quantity) into "total_quantity" }
chipoGroupByName.head()
```

```kotlin
val maxOrders = chipoGroupByName.max().get { total_quantity }
maxOrders
```

```kotlin
chipoGroupByName.filter { total_quantity == maxOrders }.get { item_name }
```
